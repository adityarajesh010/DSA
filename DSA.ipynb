```python
# station master problem #
n= len(arr)
arr= [900,940,950,1500,1800]
dep = [910,1200,1130,1900,2000]

i=0
j=0
platform= 0
maxPlatform=0
while i<n and j<n:
    if arr [i] < dep [j]:
       platform+=1
       if platform > maxPlatform:
           maxPlatform= platform
       i+=1
    else:
        platform-=1
        j+=1
print(f"the max platform: {maxPlatform}")
```

```python
""" QUESTION-4  {15,3,2,1,6,11,8} ->10
                {7,4,9,1,3,2}  -> 5   WAP to accept positive intergers list of gold prices and print maximum profit that
                 can be made with 1 byte transaction and with 1 self transaction
                 Sol: using 2 pointer """

goldprices=[7,4,9,1,3,2]
arrlen= len (goldprices)

minValue = goldprices[0]
maxprofit=0

for i in range(arrlen):
  if goldprices[i] < minValue:
    minValue = goldprices[i]
  profit = goldprices[i] - minValue
  if profit > maxprofit:
    maxprofit = profit
print(f"max profit: {maxprofit} ")
```

```python
""" QUESTION-7 IMPORTANT {([])} -------> Balanced
                         ([)] --------> Not Balanced
                         (){}[] -------> Balanced

{([])}
size 5
['','','','','']
['{','(','[','','']
top = 2                         """

paranthesis="{([])}"

stack = ['' for i in range (5)]
top =-1

for ch in paranthesis:
  if ch == '{' or ch == '[' or ch == '(':
    top+=1
    stack[top] = ch

  else:

    if stack[top] == '{' and ch == '}':
      top-=1

    if stack[top] == '[' and ch == ']':
      top-=1

    if stack[top] == '(' and ch == ')':
      top-=1
print(f"top: {top}")
```
