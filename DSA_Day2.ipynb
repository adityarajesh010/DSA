```python
""" QUESTION-1 Kdane's algorithm: Static Sliding Window Process

{7,3,1,2,4,5}
k=3
7,3,1 --> 11
3,1,2 | in : 2 out : 7 ---> 11 + 2 -7 =6
1,2,4 | in : 4 out : 3 ---> 6 + 4 -3 = 7
2,4,5 | in : 5 out : 1 ---> 7 + 5 - 1 = 11

Max sub Array Sum : 11

Step 3
Max Sub Array Sum : 11 """

arr= [7,3,1,2,4,5]
k=3
arrlen = len(arr)
# step 1

windowSum=0
maxSum = 0
for i in range(k):
    windowSum += arr[i]
maxSum = windowSum

# Step 2: Sliding Window Process

for j in range(k,arrlen):
    windowSum+= arr[j] - arr[j-k]
    if windowSum > maxSum:
        maxSum = windowSum


# Step 3:

print(f"Max Sum: {maxSum}")
```

```python
""" QUESTION-2 move all zeroes to the right side of the array withhout changing the order of the
non zero numbers.
try it wrt bubble sort

{8,0,2,1,0,3,0,6} """

arr = [8,0,2,1,0,3,0,6]
arrlen = len (arr)

for i in range(arrlen):
    for j in range(arrlen -1 - i):
      if arr[j] == 0:
        temp = arr[j]
        arr[j] = arr[j + 1]
        arr[j + 1] = temp
print(arr)
```

```python
""" QUESTION-3 {5,1,3,2,7,5}
              {1,2,3,5,5,7}
              {2,1,100,3,4,101,10,102}
    Longest Sequence Count : 4
tip: try doing it with selecetion sort
{1,2,3,4,10,100,101,102}               """

arr = [2,1,100,3,4,101,10,102]
arrlen = len(arr)

# Step 1: Sorting : Selection Sort

for i in range(arrlen):
  minIndex = i
  for j in range (i + 1, arrlen):
     if arr[j] < arr[minIndex]:
        minIndex = j
  temp = arr[i]
  arr[i] = arr[minIndex]
  arr[minIndex] = temp

# Step-2:
seqCount = 1
maxSeqCount = 0
for x in range(arrlen - 1):
   next = arr[x] + 1
   if next == arr[x + 1]:
     seqCount+= 1
   else:
      if seqCount > maxSeqCount:
        maxSeqCount = seqCount
        seqCount = 1
print(f" max sequence count: {maxSeqCount}")
```
